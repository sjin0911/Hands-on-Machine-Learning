선형적으로 분리되는 데이터셋에 LinearSVC를 훈련하고 
같은 데이터셋에 SVC와 SGDCLassifier를 적용하기

```python
from sklearn import datasets

#붓꽃 데이터셋 사용하기

iris=datasets.load_iris(as_frame=True)
X=iris.data[["petal length (cm)", "petal width (cm)"]].values
y=iris.target

setosa_or_versicolor=(y==0)|(y==1)
X=X[setosa_or_versicolor]
y=y[setosa_or_versicolor]
```

- LinearSVC는 loss="squared_hinge"를 사용하므로 세 모델 모두 loss="hinge"로 설정
- SVC 클래스는 기본적으로 RBF 커널을 사용하므로 다른 두 모델과 유사한 결과를 얻으려면 kernel="linear"로 지정
- SGDClassifier 클래스에는 C 대신 alpha라는 규제 하이퍼파라미터를 사용

```python
from sklearn.preprocessing import StandardScaler

C=5
alpha=0.05

scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
#LinearSVC
from sklearn.svm import LinearSVC

svm_clf=LinearSVC(C=C, random_state=42, loss="hinge", dual=True)
```

```python
svm_clf.fit(X_scaled,y)
```

```python
#SVC
from sklearn.svm import SVC

kernel_svm_clf=SVC(kernel="linear",C=C)
kernel_svm_clf.fit(X_scaled,y)
```

```python
#SGDClassifier
from sklearn.linear_model import SGDClassifier

sgd_clf=SGDClassifier(alpha=alpha, random_state=42)
sgd_clf.fit(X_scaled,y)
```

```python
#세 가지 모델의 결정 경계 그리기
import matplotlib.pyplot as plt

def compute_decision_boundary(model):
    w=-model.coef_[0,0]/model.coef_[0,1]
    b=-model.intercept_[0]/model.coef_[0,1]
    return scaler.inverse_transform([[-10,-10*w+b],[10,10*w+b]])

lin_line=compute_decision_boundary(svm_clf)
svc_line=compute_decision_boundary(kernel_svm_clf)
sgd_line=compute_decision_boundary(sgd_clf)

plt.figure(figsize=(11,4))
plt.plot(lin_line[:,0],lin_line[:,1],"k:", label="LinearSVC")
plt.plot(svc_line[:,0],svc_line[:,1],"b--", linewidth=2, label="SVC")
plt.plot(sgd_line[:,0],sgd_line[:,1],"r-", label="SGDClassifier")
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="Iris versicolor")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="Iris setosa")

plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.legend(loc="upper center")
plt.axis([0, 5.5, 0, 2])
plt.grid()

plt.show()
```
