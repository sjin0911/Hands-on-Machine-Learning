## k-평균 알고리즘

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import numpy as np

#무작위 숫자
blob_centers=np.array([[0.2, 2.3], [-1.5, 2.3], [-2.8, 1.8],
                      [-2.8, 2.8], [-2.8, 1.3]])
blob_std=np.array([0.4, 0.3, 0.1, 0.1, 0.1])
X, y=make_blobs(n_samples=2000, centers=blob_centers, cluster_std=blob_std,
               random_state=7)

k=5
kmeans=KMeans(n_clusters=k, n_init=10, random_state=42)
y_pred=kmeans.fit_predict(X)
```

```python
y_pred
```

```python
#레이블 찾기
y_pred is kmeans.labels_
```

```python
#센트로이드 확인
kmeans.cluster_centers_
```

```python
X_new=np.array([[0, 2], [3, 2], [-3, 3], [-3, 2.5]])
kmeans.predict(X_new)
```

```python
#각 센트로이드까지의 거리
kmeans.transform(X_new).round(2)
```

```python
#센트로이드 근사 위치를 저장한 배열로 센트로이드 초기화하기
good_init=np.array([[-3, 3],[-3, 2], [-3, 1], [-1, 2], [0, 2]])
kmeans=KMeans(n_clusters=5, init=good_init, n_init=1, random_state=42)
kmeans.fit(X)
```

```python
#모델의 이너셔 확인하기
kmeans.inertia_
```

```python
#이너셔가 작을수록 좋은 모델이기 때문에 예측기의 점수는 이너셔의 음수값(사이킷런의 규칙때문)
kmeans.score(X)
```

```python
#미니배치 k-means 알고리즘
from sklearn.cluster import MiniBatchKMeans

minibatch_kmeans=MiniBatchKMeans(n_clusters=5, n_init='auto', random_state=42)
minibatch_kmeans.fit(X)
```

```python
#실루엣 점수 계산하기
from sklearn.metrics import silhouette_score

silhouette_score(X, kmeans.labels_)
```

# 군집을 활용한 준지도 학습

```python
from sklearn.datasets import load_digits

X_digits, y_digits=load_digits(return_X_y=True)
X_train, y_train= X_digits[:1400], y_digits[:1400]
X_test, y_test=X_digits[1400:], y_digits[1400:]
```

```python
#50개의 샘플에 대한 레이블이 있다고 가정하고
#기준 성능을 얻기 위해 레이블이 있는 데이터에 대해 로지스틱 회귀 모델 훈련
from sklearn.linear_model import LogisticRegression

n_labeled=50
log_reg=LogisticRegression(max_iter=10_000)
log_reg.fit(X_train[:n_labeled], y_train[:n_labeled])
```

```python
#테스트 세트로 모델의 정확도 측정
log_reg.score(X_test, y_test)
```

```python
#KMeans
k=50
kmeans=KMeans(n_clusters=k, n_init='auto', random_state=42)
X_digits_dist=kmeans.fit_transform(X_train)

#대표이미지(각 센트로이드에서 가장 가까운 이미지) 찾기
representative_digit_idx=np.argmin(X_digits_dist, axis=0)
X_representative_digits=X_train[representative_digit_idx]
```

```python
#이미지 확인하기
plt.figure(figsize=(8,2))

for index, X_representative_digit in enumerate(X_representative_digits):
    plt.subplot(k//10, 10, index+1)
    plt.imshow(X_representative_digit.reshape(8,8), cmap="binary",
              interpolation="bilinear")
    plt.axis('off')
plt.show()
```

```python
#이미지를 보고 수동으로 레이블 할당하기
y_representative_digits=np.array([
    1, 3, 7, 0, 1, 6, 5, 6, 2, 4,
    3, 2, 9, 8, 5, 4, 1, 3, 8, 9,
    8, 4, 1, 0, 7, 7, 6, 5, 9, 3,
    4, 7, 1, 5, 2, 7, 8, 7, 3, 4,
    8, 2, 8, 2, 5, 6, 8, 9, 2, 6
])
```

```python
log_reg=LogisticRegression(max_iter=10_000)
log_reg.fit(X_representative_digits, y_representative_digits)
log_reg.score(X_test, y_test)
```

```python
#레이블 전파
y_train_propagated=np.empty(len(X_train), dtype=np.int64)
for i in range(k):
    y_train_propagated[kmeans.labels_==i]=y_representative_digits[i]
```

```python
#모델 훈련하고 성능 확인
log_reg=LogisticRegression(max_iter=10_000)
log_reg.fit(X_train, y_train_propagated)
log_reg.score(X_test, y_test)
```

```python
percentile_closest=99

X_cluster_dist=X_digits_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(k):
    in_cluster=(kmeans.labels_==i)
    cluster_dist=X_cluster_dist[in_cluster]
    cutoff_distance=np.percentile(cluster_dist, percentile_closest)
    above_cutoff=(X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff]=-1
    
partially_propagated=(X_cluster_dist != -1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
#모델 다시 훈련
log_reg= LogisticRegression(max_iter=10_000)
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
log_reg.score(X_test, y_test)
```

```python
(y_train_partially_propagated == y_train[partially_propagated]).mean()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
```

```python
#모든 샘플의 레이블
#레이블이 -1 = 이상치
dbscan.labels_
```

```python
#핵심 샘플의 인덱스
len(dbscan.core_sample_indices_)
```

```python
dbscan.core_sample_indices_
```

```python
#핵심 샘플 자체
dbscan.components_
```

```python
dbscan2=DBSCAN(eps=0.2)
dbscan2.fit(X)
```

```python
#dbscan 클래스는 predict() 메서드를 제공하지 않고 fit_predict() 메서드를 제공
#새로운 샘플에 대해 클러스터를 예측 불가능 -> 직접 예측기 선택 필요
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan2.components_, dbscan2.labels_[dbscan2.core_sample_indices_])
```

```python
#예측 만들기
X_new=np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2,1]])
knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

```python
#kneighbors를 통해 특정 샘플에 대해 훈련 세트에서 가장 가까운 k개의 이웃의 거리와 인덱스 찾기
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)
y_pred = dbscan2.labels_[dbscan2.core_sample_indices_][y_pred_idx]
y_pred[y_dist > 0.2]=-1
y_pred.ravel()
```

```python
def learned_parameters(estimator):
    return [attrib for attrib in dir(estimator)
           if attrib.endswith("_") and not attrib.startswith("_")]
```
